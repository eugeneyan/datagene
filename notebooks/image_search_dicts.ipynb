```python
import os
from logger import logger
from collections import defaultdict
import pandas as pd
```

```python
train_top_level_dir = '../data/images_sample/train_top_level/'
```

```python
index_path_dict = defaultdict()
index = 0

for image_dir in os.listdir(train_top_level_dir):
    if not image_dir.startswith('.'):  # Remove .DS_Store
        image_names = os.listdir(os.path.join(train_top_level_dir, image_dir))

        for image_name in image_names:
            if not image_name.startswith('.'):  # Remove .DS_Store
                index_path_dict[index] = image_name.split('.')[0]
                index += 1

logger.info('Index path dictionary created with {} entries'.format(len(index_path_dict)))
```

```python
index_path_dict
```

```python
asin_path_dict = defaultdict()

for image_dir in os.listdir(train_top_level_dir):
    if not image_dir.startswith('.'):  # Remove .DS_Store
        image_names = os.listdir(os.path.join(train_top_level_dir, image_dir))

        for image_name in image_names:
            if not image_name.startswith('.'):  # Remove .DS_Store
                asin = image_name.split('.')[0]
                image_path = os.path.join(train_top_level_dir, image_dir, image_name)
                asin_path_dict[asin] = image_path

logger.info('Asin path dictionary created with {} entries'.format(len(index_path_dict)))
```

```python
asin_path_dict
```

```python
category_index_dict = defaultdict()
index = 0

for image_dir in os.listdir(train_top_level_dir):
    if not image_dir.startswith('.'):
        image_names = os.listdir(os.path.join(train_top_level_dir, image_dir))
        image_count = len([img for img in image_names if not img.startswith('.')])

        category_index_dict[image_dir] = (index, image_count)
        index += 1

logger.info('Category index dictionary created with {} entries'.format(len(category_index_dict)))
```

```python
index_path_filter_dict = defaultdict()

for image_dir in os.listdir(train_top_level_dir):
    if not image_dir.startswith('.'):
        image_dir_index = category_index_dict[image_dir][0]

        index_path_filter_dict[image_dir_index] = defaultdict()
        image_index = 0

        image_names = os.listdir(os.path.join(train_top_level_dir, image_dir))

        for image_name in image_names:
            if not image_name.startswith('.'):
                index_path_filter_dict[image_dir_index][image_index] = image_name.split('.')[0]
                image_index += 1

logger.info('Index path filter dictionary created with {} entries'.format(len(index_path_filter_dict)))
```

```python
index_path_filter_dict
```

```python
title_df = pd.read_csv('../data/metadata.csv', usecols=['asin', 'title'])
title_df = title_df.iloc[-1000000:, :]
```

```python
category_df = pd.read_csv('../data/image_category.csv', nrows=3000000, usecols=['asin', 'category_path'])
```

```python
df = title_df.merge(category_df)
```

```python
df['title'].fillna('TITLE MISSING', inplace=True)
df['category_path'].fillna('CATEGORY MISSING', inplace=True)
```

```python
valid_images = index_path_dict.values()
valid_images = [img.split('.')[0] for img in valid_images]
```

```python
df = df[df['asin'].isin(valid_images)]
```

```python
df
```

```python
asin_dict = defaultdict()

for index, row in df.iterrows():
    asin = row['asin']
    image_path = index_path_dict[asin]
    title = row['title']
    category = row['category_path']
    
    asin_dict[asin] = (image_path, title, category)
```
